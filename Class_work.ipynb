-  investigate outliers in the data through Boxplot

```python


# Setting up the matplotlib figure
plt.figure(figsize=(15, 10))

# Plotting boxplots to visualize outliers for each feature
colors = ['red', 'blue', 'green', 'yellow']
for i, column in enumerate(df.columns, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(y=df[column], color=colors[i%4])
    plt.title(column)

plt.tight_layout()
plt.show()
```
